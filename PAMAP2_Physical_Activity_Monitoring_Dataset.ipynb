```python
import pandas as pd

# Correct path after double unzip
file_path = "pamap2_dataset/PAMAP2_Dataset/Protocol/subject101.dat"

cols = [
    "timestamp", "activity_id", "heart_rate",
    "IMU_hand_temp", "IMU_hand_acc16_x", "IMU_hand_acc16_y", "IMU_hand_acc16_z",
    "IMU_hand_acc6_x", "IMU_hand_acc6_y", "IMU_hand_acc6_z",
    "IMU_hand_gyro_x", "IMU_hand_gyro_y", "IMU_hand_gyro_z",
    "IMU_hand_mag_x", "IMU_hand_mag_y", "IMU_hand_mag_z",
    "IMU_hand_ori1", "IMU_hand_ori2", "IMU_hand_ori3", "IMU_hand_ori4",
    "IMU_chest_temp", "IMU_chest_acc16_x", "IMU_chest_acc16_y", "IMU_chest_acc16_z",
    "IMU_chest_acc6_x", "IMU_chest_acc6_y", "IMU_chest_acc6_z",
    "IMU_chest_gyro_x", "IMU_chest_gyro_y", "IMU_chest_gyro_z",
    "IMU_chest_mag_x", "IMU_chest_mag_y", "IMU_chest_mag_z",
    "IMU_chest_ori1", "IMU_chest_ori2", "IMU_chest_ori3", "IMU_chest_ori4",
    "IMU_ankle_temp", "IMU_ankle_acc16_x", "IMU_ankle_acc16_y", "IMU_ankle_acc16_z",
    "IMU_ankle_acc6_x", "IMU_ankle_acc6_y", "IMU_ankle_acc6_z",
    "IMU_ankle_gyro_x", "IMU_ankle_gyro_y", "IMU_ankle_gyro_z",
    "IMU_ankle_mag_x", "IMU_ankle_mag_y", "IMU_ankle_mag_z",
    "IMU_ankle_ori1", "IMU_ankle_ori2", "IMU_ankle_ori3", "IMU_ankle_ori4"
]

# Load subject101 data
data = pd.read_csv(file_path, sep=" ", header=None, names=cols)

print("Shape:", data.shape)
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# --- Label distribution ---
plt.figure(figsize=(12,6))
sns.countplot(x="activity_id", data=data, palette="tab20")
plt.title("Activity Distribution (Raw IDs)")
plt.show()

# --- Missing values ---
plt.figure(figsize=(10,4))
sns.heatmap(data.isnull(), cbar=False)
plt.title("Missing Value Heatmap")
plt.show()

print("Total missing values:", data.isna().sum().sum())

# --- Histograms of selected features ---
sample_feats = ["heart_rate", "IMU_hand_acc16_x", "IMU_chest_gyro_y", "IMU_ankle_mag_z"]
data[sample_feats].hist(figsize=(12,8), bins=50)
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()

# --- Correlation heatmap ---
plt.figure(figsize=(8,6))
sns.heatmap(data[sample_feats].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap (Selected Features)")
plt.show()

# --- Boxplots across activities ---
plt.figure(figsize=(14,6))
sns.boxplot(x="activity_id", y="heart_rate", data=data.sample(5000), palette="Set2")
plt.title("Heart Rate Distribution Across Activities")
plt.xticks(rotation=90)
plt.show()

# --- Time-series plot (sample signals) ---
plt.figure(figsize=(14,6))
plt.plot(data["heart_rate"][:1000], label="Heart Rate")
plt.plot(data["IMU_hand_acc16_x"][:1000], label="Hand Acc X")
plt.legend()
plt.title("Sample Time-Series Signals")
plt.show()

# --- PCA visualization ---
X = data.drop(["timestamp", "activity_id"], axis=1).fillna(0)
y = data["activity_id"]

X_scaled = StandardScaler().fit_transform(X.sample(5000, random_state=42))
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(10,6))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=y.sample(5000, random_state=42), palette="tab20", s=20, alpha=0.7)
plt.title("PCA Projection of PAMAP2 Sensor Data (Sampled)")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# --- Handle missing values ---
data_clean = data.fillna(method="ffill").fillna(method="bfill")

# --- Features & labels ---
X = data_clean.drop(["timestamp", "activity_id"], axis=1)
y = data_clean["activity_id"]

# --- Normalize features ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- Encode activity labels ---
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# --- Train-test split ---
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

print("Train shape:", X_train.shape, "Test shape:", X_test.shape)
```
